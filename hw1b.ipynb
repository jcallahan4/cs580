## Problem 4
Compute the condition number for the following matrix $A$:


$\begin{align}\begin{bmatrix}
  8 & 6 & 10 & 10 \\
      9 & 1 & 10 & 5 \\
      1 & 3 & 2 & 8 \\
      9 & 5 & 10 & 1
 \end{bmatrix}\end{align}$
 
 ### Solution
 The condition number is given by the ratio between the maximum and minimum singular values:

```python
A = np.array([[8, 6, 10, 10],
              [9, 1, 10, 5],
              [1, 3, 2, 8],
              [9, 5, 10, 1]])

U, σ, V = np.linalg.svd(A)
condition = np.max(σ)/np.min(σ)
print(f'The condition number of A is {condition}')
```

### Part c)

```python
def show_approx(U, sig, V, n):
    sig_n = np.pad(sig[:n], (0,len(sig)-n))

    Sig_n = np.zeros((U.shape[1], V.shape[0]))
    Sig_n[:len(sig_n), :len(sig_n)] = np.diag(sig_n)
    A_n = U @ Sig_n @ V
    
    plt.imshow(A_n, cmap='gray')
    plt.axis('off')
    plt.title(f'Reconstruction with n={n}')
    plt.show()
```

```python
ns = [10, 20, 40]
for n in ns:
    show_approx(U,sig,V,n)
```

How many terms $n$ do you do need to retain in the sum to recognize that the approximation $A_n$ is a face?

```python
ns = np.arange(10)
for n in ns:
    show_approx(U,sig,V,n)
```
